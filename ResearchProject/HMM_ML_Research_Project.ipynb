# Train a HMM for Expert at Task 2

```python
# Train a HMM for the Experts in Task 2

states = ["Expert", "Novice"]
n_states = len(states)

df_U3T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Experts/U3T2.csv')
df_U4T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Experts/U4T2.csv')
df_U5T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Experts/U5T2.csv')
df_U6T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Experts/U6T2.csv')
df_U7T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Experts/U7T2.csv')
df_U8T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Experts/U8T2.csv')
# Rainy Season: Observation sequence for Umbrellas for 30 days in a Rainy season
X_train_U3T2 = np.vstack((df_U3T2['duration'], df_U3T2['line_type'])).T
X_train_U4T2 = np.vstack((df_U4T2['duration'], df_U4T2['line_type'])).T
X_train_U5T2 = np.vstack((df_U5T2['duration'], df_U5T2['line_type'])).T
X_train_U6T2 = np.vstack((df_U6T2['duration'], df_U6T2['line_type'])).T
X_train_U7T2 = np.vstack((df_U7T2['duration'], df_U7T2['line_type'])).T
X_train_U8T2 = np.vstack((df_U8T2['duration'], df_U8T2['line_type'])).T

print(X_train_U3T2.shape)
X_train_expert = np.concatenate([X_train_U3T2,X_train_U4T2 ,X_train_U5T2, X_train_U6T2,X_train_U7T2,X_train_U8T2]).T
lengths = [len(X_train_U3T2), len(X_train_U4T2), len(X_train_U5T2), len(X_train_U6T2), len(X_train_U7T2), len(X_train_U8T2)]

# Learning model parameters: transmission & emission matrices

    
model_train_experts_task2= hmm.GaussianHMM(n_components=n_states, covariance_type="full", n_iter=1000)
model_train_experts_task2.fit(X_train_expert.reshape(-1,1), lengths)
    
    
print("Expert HMM: Transition Probabilities: ")
print(model_train_experts_task2.transmat_)

print("Expert HMM start prob:", model_train_experts_task2.startprob_)
```

# Train a HMM for Novices at Task 2

```python


states = ["Expert", "Novice"]
n_states = len(states)

df_U23T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Novices/U23T2.csv')
df_U24T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Novices/U24T2.csv')
df_U25T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Novices/U25T2.csv')
df_U26T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Novices/U26T2.csv')
df_U27T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Novices/U27T2.csv')
df_U28T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_training+'/Task2/Novices/U28T2.csv')
# Rainy Season: Observation sequence for Umbrellas for 30 days in a Rainy season
X_train_U23T2 = np.vstack((df_U23T2['duration'], df_U23T2['line_type'])).T
X_train_U24T2 = np.vstack((df_U24T2['duration'], df_U24T2['line_type'])).T
X_train_U25T2 = np.vstack((df_U25T2['duration'], df_U25T2['line_type'])).T
X_train_U26T2 = np.vstack((df_U26T2['duration'], df_U26T2['line_type'])).T
X_train_U27T2 = np.vstack((df_U27T2['duration'], df_U27T2['line_type'])).T
X_train_U28T2 = np.vstack((df_U28T2['duration'], df_U28T2['line_type'])).T

X_train_novices = np.concatenate([X_train_U23T2,X_train_U24T2 ,X_train_U25T2, X_train_U26T2, X_train_U27T2, X_train_U28T2]).T
lengths_novices = [len(X_train_U23T2), len(X_train_U24T2), len(X_train_U25T2), len(X_train_U26T2), len(X_train_U27T2), len(X_train_U28T2)]

# Learning model parameters: transmission & emission matrices

    
model_train_novices_task2= hmm.GaussianHMM(n_components=n_states,covariance_type="full", n_iter=1000)
model_train_novices_task2.fit(X_train_novices.reshape(-1,1), lengths_novices)
    
    
print("Novice HMM: Transition Probabilities: ")
print(model_train_novices_task2.transmat_)
print("Novice start prob:", model_train_novices_task2.startprob_)
```

# Evaluation of Expert Participant on both the HMMs

```python
print("\n--------------------------------- Model: Expert -----------------------------------")
print("Testing with Expert User")
model_test_expert = hmm.GaussianHMM(n_components=n_states, init_params="")


# Rainy
model_test_expert.startprob_ = model_train_experts_task2.startprob_

model_test_expert.transprob_ = np.array([
  [model_train_experts_task2.transmat_[0, 0], model_train_experts_task2.transmat_[0, 1]],
  [model_train_experts_task2.transmat_[1, 0], model_train_experts_task2.transmat_[1, 1]]
])



df_U1T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_test+'/Task2/Experts/U1T2.csv')
X_test_U1T2 = np.array([df_U1T2['line_type']]).T
# Estimate model parameters.
model_test_expert = model_test_expert.fit(X_test_U1T2.reshape(-1,1))


# Find most likely state (hidden) sequence corresponding to the observation sequence
# We use the following function "decode": the first argument should be the observation sequence (matrix)
logprob_expert_task2, predicted_hidden_states_rainy = model_test_expert.decode(X_test_U1T2, algorithm="viterbi")




print("\nProbability of Expert at Task 2:")
print(logprob_expert_task2)


print("\n--------------------------------- Model: Novice -----------------------------------")



model_test_novice = hmm.GaussianHMM(n_components=n_states, init_params="")


# Rainy
model_test_novice.startprob_ = model_train_novices_task2.startprob_

model_test_novice.transprob_ = np.array([
  [model_train_novices_task2.transmat_[0, 0], model_train_novices_task2.transmat_[0, 1]],
  [model_train_novices_task2.transmat_[1, 0], model_train_novices_task2.transmat_[1, 1]]
])



# df_U22T2= pd.read_csv(CONSTANTS.path_to_csv_per_user_test+'/Task2/Novices/U22T2.csv')
# X_test_U22T2 = np.array([df_U22T2['line_type']]).T
# Estimate model parameters.
model_test_novice = model_test_novice.fit(X_test_U1T2.reshape(-1,1))


# Find most likely state (hidden) sequence corresponding to the observation sequence
# We use the following function "decode": the first argument should be the observation sequence (matrix)
logprob_novice_task2, predicted_hidden_states_rainy = model_test_novice.decode(X_test_U1T2, algorithm="viterbi")




print("\nProbability of Novice at Task 2:")
print(logprob_novice_task2)


if logprob_expert_task2 > logprob_novice_task2:
    print("Participant is an Expert")
    
else:
    print("Participant is an Novice")
```
