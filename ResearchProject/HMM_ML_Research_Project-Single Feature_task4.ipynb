# Train a HMM for Experts at Task 2 

```python
# Train a HMM for the Experts in Task 2

states = ["Expert", "Novice"]
n_states = len(states)

df_U6T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Experts/U6T4_new.csv')
df_U25T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Experts/U25T4_new.csv')
df_U26T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Experts/U26T4_new.csv')
df_U27T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Experts/U27T4_new.csv')
df_U30T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Experts/U30T4_new.csv')


X_train_U6T4 = np.array((df_U6T4['change_line']))
X_train_U25T4 = np.array((df_U25T4['change_line']))
X_train_U26T4 = np.array((df_U26T4['change_line']))
X_train_U27T4 = np.array((df_U27T4['change_line']))
X_train_U30T4 = np.array((df_U30T4['change_line']))


X_train_expert = np.concatenate([X_train_U6T4, X_train_U25T4,X_train_U26T4,X_train_U27T4,X_train_U30T4]).T
lengths = [len(X_train_U6T4), len(X_train_U25T4), len(X_train_U26T4), len(X_train_U27T4), len(X_train_U30T4)]

# Learning model parameters: transmission & emission matrices

    
model_train_experts_task2= hmm.GaussianHMM(n_components=n_states, covariance_type="full", n_iter=1000)
model_train_experts_task2.fit(X_train_expert.reshape(-1,1), lengths)
    
    
print("Expert HMM: Transition Probabilities: ")
print(model_train_experts_task2.transmat_)

print("Expert HMM start prob:", model_train_experts_task2.startprob_)
```

# Train a HMM for Novices at Task 2

```python


states = ["Expert", "Novice"]
n_states = len(states)

df_U5T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Novices/U5T4_new.csv')
df_U11T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Novices/U11T4_new.csv')
df_U13T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Novices/U13T4_new.csv')
df_U23T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Novices/U23T4_new.csv')
df_U28T4= pd.read_csv(CONSTANTS.path_to_csv_per_user_correctness_training+'/Task4/Novices/U28T4_new.csv')



X_train_U5T4 = np.array((df_U5T4['change_line']))
X_train_U11T4 = np.array((df_U11T4['change_line']))
X_train_U13T4 = np.array((df_U13T4['change_line']))
X_train_U23T4 = np.array((df_U23T4['change_line']))
X_train_U28T4 = np.array((df_U28T4['change_line']))

X_train_novices = np.concatenate([X_train_U5T4,X_train_U23T4 , X_train_U11T4, X_train_U11T4, X_train_U28T4]).T
lengths_novices = [len(X_train_U5T4), len(X_train_U23T4), len(X_train_U11T4), len(X_train_U11T4), len(X_train_U28T4)]

# Learning model parameters: transmission & emission matrices


model_train_novices_task2= hmm.GaussianHMM(n_components=n_states,covariance_type="full", n_iter=1000)
model_train_novices_task2.fit(X_train_novices.reshape(-1,1), lengths_novices)
    
    
print("Novice HMM: Transition Probabilities: ")
print(model_train_novices_task2.transmat_)
print("Novice start prob:", model_train_novices_task2.startprob_)
```
