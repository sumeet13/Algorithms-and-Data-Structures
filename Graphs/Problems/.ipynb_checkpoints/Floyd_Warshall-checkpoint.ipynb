```python
import sys
INF = float('-inf')


def initialize():

    for i in range(V):
        for j in range(V):

            dist[i][j] = graph[i][j]

            if dist[i][j] == INF:
                next_[i][j] = -1

            else:
                next_[i][j] = j


def floydWarshall():

    for k in range(V):

        # pick all vertices as source one by one
        for i in range(V):

            for j in range(V):

                # If vertex k is on the shortest path from
                # i to j, then update the value of dist[i][j]

                if dist[i][j] < dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    next_[i][j] = next_[i][k]
    printSolution()


# A utility function to print the solution
def printSolution():
    print(
        "\nFollowing matrix shows the shortest distances between every pair of vertices"
    )

    for i in range(V):
        print(dist[i])

    print("\nFollowing matrix shows the path matrix")
    for i in range(V):
        print(next_[i])


def constructPath(u, v):
    path = []

    print("\nThe path between from %d -> %d is: " % (u, v))
    if next_[u][v] == -1:
        print("No path exists")
        return
    path.append(u)
    while (u != v):
        u = next_[u][v]
        path.append(u)

    print(path)


def findCriticalPath():
    u, v = -1, -1
    max_timing = INF
    for i in range(V):
        for j in range(V):

            if max_timing < dist[i][j]:

                max_timing = dist[i][j]
                u, v = i, j
    return u, v


# Driver program to test the above program
# Let us create the following weighted graph
""" 
            10 
       (0)------->(3) 
        |         /|\ 
      5 |          | 
        |          | 1 
       \|/         | 
       (1)------->(2) 
            3           """
graph = [[0, 5, INF, 10], 
        [INF, 0, 3, INF], 
        [INF, INF, 0, 1],
        [INF, INF, INF, 0]]
# Print the solution
V = len(graph)

dist = [[0 for j in range(V)] for _ in range(V)]

next_ = [[-1 for j in range(V)] for _ in range(V)]

initialize()
floydWarshall()

u, v = findCriticalPath()

constructPath(u, v)
```
