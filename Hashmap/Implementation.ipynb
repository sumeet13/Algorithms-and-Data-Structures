```python
class HashMap():
    def __init__(self):
        self.INITIAL_CAPACITY = 10
        self.slots = [None] * self.INITIAL_CAPACITY

    class Node():
        def __init__(self, key, value):
            self.key = key
            self.value = value
            self.next = None

        def set_next(self, next_node):
            self.next = next_node

        def get_next(self):
            return self.next

        def set_value(self, value):
            self.value = value

        def get_value(self):
            return self.value

        def set_key(self, key):
            self.next = next_node

        def get_key(self):
            return self.key

    def hash_fun(self, key):
        hash_sum = ord(str(key)) % self.INITIAL_CAPACITY

        return hash_sum

    def put(self, key, data):
        index = self.hash_fun(key)
        print("index", index)
        if self.slots[index] == None:
            print("43----")
            self.slots[index] = self.Node(key, data)

        else:
            # there is a collision

            prev = node = self.slots[index]
            print("50----")
            while node != None:
                # Either you reach till the end of the linkedlist or you found a key to replace

                if node.get_key() == key:
                    print("matched")
                    # replace the value
                    node.set_value(data)
                    return

                prev = node
                node = node.get_next()

            prev.set_next(self.Node(key, data))

    def get(self, key):
        index = self.hash_fun(key)

        if not self.slots[index]:
            raise KeyError("Key does not exist")

        else:
            # there is a collision

            prev = node = self.slots[index]

            while node != None:
                # Either you reach till the end of the linkedlist or you found a key to replace

                if node.get_key() == key:
                    # replace the value
                    return node.get_value()

                prev = node
                node = node.get_next()

            if prev.get_next() == None:
                raise KeyError("Key does not exist")

    def remove(self, key):
        index = self.hash_fun(key)
        if self.slots[index].get_key() == key:
            node = self.slots[index]
            self.slots[index] = node.get_next()
            node.set_next(None)

        else:
            # there is a collision
            prev = None
            node = self.slots[index]

            while node.get_next() != None or node.get_key() != key:
                # Either you reach till the end of the linkedlist or you found a key to replace

                prev = node
                node = node.get_next()

            if node.get_next() == None:
                raise KeyError("Key does not exist")

            if node.get_key() == key:
                prev.set_next(node.get_next())
                node.set_next(None)


hashMap = HashMap()
hashMap.put(1, 2)
hashMap.put(1, 3)
hashMap.put(';', ';')
hashMap.put('E', 'E')
hashMap.put(1, "abc")
hashMap.put(2, 0)
hashMap.put(3, "abc")

print(hashMap.get(3))
print(hashMap.get(1))
hashMap.remove(1)
print(hashMap.get('E'))
print(hashMap.get(';'))
hashMap.get(1)
```
