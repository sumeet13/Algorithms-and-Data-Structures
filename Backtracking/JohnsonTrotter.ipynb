```python
from copy import copy
def isValidSudoku(board):

    #create a dictionary

    dic = {str(i):1  for i in range(10)}
    temp = {str(i):1  for i in range(10)}

    #validation of the columns
    for i in range(9):
        print("before cols: ",temp, dic,i)
        for j in range(9):

            if board[i][j]!=".":
                if temp[board[i][j]]:
                    temp[board[i][j]] -= 1

                else:
                    return False

        temp = dic.copy()
        print("cols: ",temp, dic,i,j)
    #validation of the rows
    for i in range(9):
        print("before rows: ",temp, dic,i,j)
        for j in range(9):

            if board[j][i]!=".":
                if temp[board[j][i]]:
                    temp[board[j][i]] -= 1

                else:
                    return False

        temp = copy(dic)
        
    #validation of the the sub matrixes
    for k in range(3):
        for l in range(3):
            
            for i in range(3):
                for j in range(3):
                    
                    if board[i+k*3][j+l*3]!=".":
                        if temp[board[i+k*3][j+l*3]]:
                            temp[board[i+k*3][j+l*3]] -= 1

                        else:
                            return False
            print("boxes: ",temp, dic,i,j,k,l)                
            
            temp = copy(dic)
            

    return True
a = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
isValidSudoku(a)
```
